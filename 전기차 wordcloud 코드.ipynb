### concat으로 합친 csv -> S

```python
import pandas as pd

# set files path
sales5 = 'C:\\Users\\ASIA-08\\Desktop\\빅데이터(PDF+코드)\\프로젝트(전기차)\\5.csv'
sales6 = 'C:\\Users\\ASIA-08\\Desktop\\빅데이터(PDF+코드)\\프로젝트(전기차)\\6.csv'
sales7 = 'C:\\Users\\ASIA-08\\Desktop\\빅데이터(PDF+코드)\\프로젝트(전기차)\\7.csv'
sales8 = 'C:\\Users\\ASIA-08\\Desktop\\빅데이터(PDF+코드)\\프로젝트(전기차)\\8.csv'
sales9 = 'C:\\Users\\ASIA-08\\Desktop\\빅데이터(PDF+코드)\\프로젝트(전기차)\\9.csv'

print("*** Merging multiple csv files into a single pandas dataframe ***")

# merge files
dataFrame = pd.concat(
   map(pd.read_csv, [sales5, sales6, sales7, sales8, sales9]), ignore_index=True)
print(dataFrame)
```

```python
dataFrame.to_csv(('C:\\Users\\ASIA-08\\Desktop\\S.csv'), sep=',', na_rep='NaN', encoding='utf-8-sig')
```

```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image

import os

import nltk

print(os.getcwd())

# 인터넷에서 받아온 텍스트파일
text = open('C:\\Users\\ASIA-08\Desktop\\빅데이터(PDF+코드)\\프로젝트(전기차)\\S.csv').read() 
alice_mask = np.array(Image.open('C:\\Users\\ASIA-08\\Desktop\\빅데이터(PDF+코드)\\프로젝트(전기차)\\사회.jpg'))

stopwords = []

f = open('C:\\Users\\ASIA-08\\Desktop\\빅데이터(PDF+코드)\\프로젝트(전기차)\\stopwords.txt')
lines = f.readlines()
for line in lines:
    line = line.strip()
    stopwords.append(line)
f.close()
print(stopwords)
nltk.download('punkt')
```

```python
import matplotlib.pyplot as plt
import platform

path = 'c:\\Windows\\Fonts\\malgun.ttf'

from matplotlib import font_manager, rc
if platform.system() == 'Darwin':
    rc('font', family = 'AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family = font_name)
else:
    print('Unknown system... sorry~~~')
    
%matplotlib inline

plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color = 'white', max_words = 
               4000, mask = alice_mask, font_path = path, stopwords = stopwords)
wc = wc.generate(text)
wc.words_

plt.figure(figsize=(10,10))
plt.imshow(wc,interpolation = 'bilinear')
plt.axis('off')
plt.show()
```
